```python
import pandas as pd
data = {
    "Name": ["Asha","Ravi","Kiran","Meera","Vikram","Sona"],
    "Age": [24,27,None,26,32,25],
    "Marks": [85,90,78,88,None,75],
    "Gender": ["Female","Male","Male","Female","Male","Female"],
    "Department": ["HR","IT","IT","Finance","HR","Finance"]
}
df=pd.DataFrame(data)
print("->Print first 5 rows using head():\n",df.head())
print("\n->Print first 5 rows using tail():\n",df.tail())
print("\n->Print structure using info():")
df.info()
print("\n->Print statistics using describe():\n",df.describe())
print("\n->Print dataset size using shape:\n",df.shape)
print("\n->Print all column names using columns:\n",df.columns)
print("\n->Print each column datatype using dtypes:\n",df.dtypes)
print("\n->Print missing values using isnull():\n",df.isnull())
print("\n->Number of missing values using isnull().sum():\n",df.isnull().sum())
```
