```python
import pandas as pd
data = {
    "EmpID": [101, 102, 103, 104, 105, 106],
    "Name": ["Arun", "Divya", "John", "Meenu", "Rahul", "Sara"],
    "Company": ["TCS", "TCS", "Infosys", "Infosys", "Wipro", "Wipro"],
    "Year": [2023, 2024, 2023, 2024, 2023, 2024],
    "Salary": [45000, 48000, None, 52000, 41000, 46000],
    "Role": ["Developer", "Tester", "Developer", "HR",None, "Tester"]
}
df=pd.DataFrame(data)
print("->Print first 3 rows:\n",df.head(3))
print("\n->Print last 2 rows:\n",df.tail(2))
print("\n->Full dataset structure using info():")
df.info()
print("\n->Statistical summary using describe():\n",df.describe())
print("\n->Print dataset size using shape:\n",df.shape)
print("\n->Print all column names using columns:\n",df.columns)
print("\n->Print each column datatype using dtypes:\n",df.dtypes)
print("\n->Print missing values using isnull():\n",df.isnull())
print("\n->Number of missing values using isnull().sum():\n",df.isnull().sum())
arrays = [df["Company"], df["Year"]]
index = pd.MultiIndex.from_arrays(arrays, names=("Company", "Year"))
df_multi = df.copy()
df_multi.index = index
print(df_multi)
print("->Print first rows using head:\n",df.head())
```
