```python
import pandas as pd
data = {
    "StudentID": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Ravi", "Kiran", "Meera", "John"],
    "Age": [24, 30, 28, 26, 22],
    "Dept": ["HR", "IT", "IT", "Finance", "IT"],
    "Marks": [78, 90, 84, 92, 65],
    "Status": ["pending", "completed", "pending", "completed", "pending"],
    "City": ["New Delhi", "Mumbai", "New York", "New Delhi", "Yorkshire"]
}
df=pd.DataFrame(data)
print(df)
print("\n->Print only Name using []:")
print(df["Name"])
print("\n->Print only Name and mark using []:")
print(df[['Name', 'Marks']])
print("\n->Print row with index 0 using loc[]:\n",df.loc[0])
print("\n->Print Print Marks of index 1 using loc[]:",df.loc[1, 'Marks'])
print("\n->Print all rows but only Name and Dept:\n",df.loc[:, ['Name', 'Dept']])
print("\n->Print 3rd row using iloc[]:\n",df.iloc[2])
print("\n->Print value at 2nd row, 5th column using iloc[]:\n",df.iloc[1,4])
print("\n->Print all rows but only Name and Marks using iloc[]:\n",df.iloc[:,[1,4]])
print("\n->Select rows 0 to 2 and columns Name and Marks:\n",df.loc[0:2, ['Name', 'Marks']])
print("\n->Slice first 3 rows using .iloc:\n",df.iloc[0:3])
print("\n->Students with Marks >= 85:\n",df[df['Marks'] >= 85])
print("\n->Students whose Dept is IT AND Age > 25:\n",df[(df["Age"] > 25) & (df["Dept"] == "IT")])
print("\n->Students where Marks >= 80 and Dept == IT:\n", df.query("Age > 25 and Dept == 'IT'"))
df["BonusMarks"] = df["Marks"] +5
df.loc[df["Status"] == "pending", "Status"] = "in progress"
df = df.drop(columns=["Dept"])
df = df.map(lambda x: x*1)
print("\n->df after applying transformations:\n",df)
filtered = df[df["City"].str.contains("York")]
print("\n->Students where City contains York:\n",filtered)
bins = [0, 70, 85, 100]
labels = ["Low", "Average", "High"]
df["Performance"] = pd.cut(df["Marks"], bins=bins, labels=labels)
print("\n->After categorizing marks:\n",df[['Name','Marks','Performance']])
```
