```python
import pandas as pd
data = {
    "StudentID": [101, 102, 101, 104, 105, 106],
    "Name": ["Asha", "Ravi", "Asha", "Kiran", "Meera", "John"],
    "City": ["mumbai", "Delhi", "mumbai", "chennai", "Kolkata", "Delhi"],
    "Marks": [85, None, 85, 78, 90, 70],
    "Grade": ["A", "B", "A", "F", "A", "B"],
    "Status": ["pending", "pending", "pending", "completed", "completed", "pending"]
}
df=pd.DataFrame(data)
df_filled=df.fillna(0)
print("->After handling missing values:\n",df_filled)
df_dropped = df.dropna()
print("\n->Rows with missing values are removed using dropna():\n",df_dropped)
print("\n->Replace Wrong Values:")
df_replaced = df_dropped.replace({'F': 'Fail'})
print(df_replaced)
df.duplicated()
df_unique=df_replaced.drop_duplicates()
print("\n->Identified and removed duplicates:\n",df_unique)
df_unique["City_Upper"] = df_unique["City"].str.upper()
print("\n->A new column City_Upper using .str.upper():\n",df_unique)
df_renamed = df_unique.rename(columns={'Marks': 'Score'})
print("->After renaming the column mark to score:\n",df_renamed)
sorted_df = df_renamed.sort_values(by='Score', ascending=False)
print("\n->After sorting using sort_values():\n",sorted_df)
df = df.sort_index()
print("\n->After sorting using sort_index():\n",df)
```
