```python
import pandas as pd
data = {
    "OrderID": [1, 2, 3, 4, 5, 6],
    "CustomerName": ["Arun", "Neha", "Arun", "Sona", "Rahul", "Maya"],
    "Product": ["Apple Juice", "Mango Shake", "Apple Juice", "Orange Juice", "Banana Smoothie", "Apple Juice"],
    "Qty": [2,None, 2, 1, 3, 1],
    "Price": [120, 150, 120,None, 200, 120],
    "City": ["kochi", "Delhi", "kochi", "Mumbai", "Chennai", "new york"],
    "Status": ["pending", "completed", "pending", "pending", "completed", "pending"]
}

df = pd.DataFrame(data)
print("->",df)
df_filled = df['Qty'].fillna(1)
df_filled = df['Price'].fillna(0)
print("\n->Filled missing values:",df_filled)
df_dropped = df.dropna()
print(df_dropped)
print("\n->After removing Bad rows:",df_filled)
df.duplicated()
df_clean = df.drop_duplicates()
print("\n->Removing Duplicates:\n:",df_clean)
df_clean["City_Upper"] = df_clean["City"].str.upper()
print("\n->AFTER CONVERTING CITY TO UPPERCASE:\n",df_clean)
juice_orders = df_clean["Product"].str.contains("Juice")
print("\n->DATAFRAME AFTER FILTERING ORDERS WITH PRODUCT CONTAINS 'JUICE':\n", juice_orders)
df_clean = df_clean.rename(columns={
    "CustomerName": "Customer",
    "Qty": "Quantity"
})
print("\n->DATAFRAME AFTER RENAMING CUSTOMERNAME AND QTY:\n",df_clean)
df_clean["Quantity"] = df_clean["Quantity"].astype(int)
df["Price"] = df["Price"].astype(float)
print("\n->DATAFRAME AFTER COVERTING QUANTITY TO INT AND PRICE TO FLOAT:\n",df) 
df_clean["Status"] = df_clean["Status"].replace("pending", "in progress")
print("\n->DATAFRAME AFTER REPLACING STATUS:\n",df_clean)
print("\n->DATAFRAME AFTER SORTING PRICE IN DESCENDING ORDER:\n") 
print(df_clean.sort_values(by="Price", ascending=False))
df_clean= df_clean.set_index("OrderID")
print("\n->DATAFRAME AFTER SETTING ORDERID AS INDEX:\n", df_clean)
df_clean = df_clean.reset_index()
print("\n->DATAFRAME AFTER RESETTING INDEX:\n",)
print(df_clean.sort_index())
print("\n->DATAFRAME AFTER SORTING BY INDEX:\n",df_clean)
```
