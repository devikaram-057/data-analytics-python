```python
import pandas as pd
student_df = pd.DataFrame({
    "StudentID": [1, 2, 3, 4],
    "Name": ["alice", "bob", "charlie", "diya"],
    "Dept": ["IT", "HR", "IT", "Finance"]
})
marks_df = pd.DataFrame({
    "StudentID": [1, 2, 3, 5],
    "Math": [80, 45, 90, 70],
    "Science": [75, 60, 88, 65]
})
inner_df = pd.merge(student_df, marks_df, on="StudentID", how="inner")
print("->DATAFRAME AFTER MERGING USING INNER JOIN:\n", inner_df)
left_df = pd.merge(student_df, marks_df, on="StudentID", how="left")
print("\n->DATFRAME AFTER MERGING USING LEFT JOIN:\n", left_df)
left_df["Total"] = left_df["Math"] + left_df["Science"]
print("\n->DATAFRAME AFTER VECTORIZATIONS:\n",marks_df)
print("\n->SUM:\n", left_df[["Math", "Science"]].sum())
print("\n->MEAN:\n", left_df[["Math", "Science"]].mean())
print("\n->MIN Total:", left_df["Total"].min())
print("->MAX Total:", left_df["Total"].max())
left_df["Result"] = left_df["Total"].apply(lambda x: "Pass" if x >= 150 else "Fail")
print("\n->DATAFRAME AFTER ADDING RESULT COLUMN:\n",left_df)
left_df["->Percentage"] = left_df.apply(lambda row: (row["Total"] / 200) * 100, axis=1)
print("\n->DATAFRAME AFTER ADDING PERCENTAGE COLUMN:\n",left_df)
left_df["Name"] = left_df["Name"].map(lambda x: x.title()) 
print("\n->DATAFRAME AFTER USING groupby  to transform mean to total:\n")
left_df["Dept_Avg_Total"] = left_df.groupby("Dept")["Total"].transform("mean")
week1_df = pd.DataFrame({
    "StudentID": [1, 2],
    "Math": [70, 60],
    "Science": [65, 55]
})
week2_df = pd.DataFrame({
    "StudentID": [3, 4],
    "Math": [80, 75],
    "Science": [78, 72]
})
print("\n->WEEK1 DATAFRAME:\n",week1_df)
print("\n->WEEK2 DATAFRAME:\n",week2_df)
combined_df = pd.concat([week1_df, week2_df], axis=0)
print("\n->COMBINED WEEKLY DATA:\n", combined_df)
print("\n->FINAL REPORT:\n",left_df)
```
