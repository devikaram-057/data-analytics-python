```python
import pandas as pd
emp_df = pd.DataFrame({
    "EmpID": [101, 102, 103, 104],
    "Name": ["asha", "ravi", "kiran", "meera"],
    "Dept": ["IT", "HR", "IT", "Finance"]
})
salary_df = pd.DataFrame({
    "EmpID": [101, 102, 103, 105],
    "Salary": [50000, 40000, 55000, 45000]
})
bonus_df = pd.DataFrame({
    "EmpID": [101, 102, 104],
    "Bonus": [5000, 3000, 4000]
})
print("\n->EMPLOYEE DATAFRAME:\n",emp_df)
print("\n->SALARY DATAFRAME:\n",salary_df)
print("\n->BONUS DATAFRAME:\n",bonus_df)
df = pd.merge(emp_df,salary_df, on="EmpID", how="left")
df = pd.merge(df, bonus_df, on="EmpID", how="outer")
print("\n->JOINED DATAFRAME AFTER CONCATINATION:\n",df)
df["Salary"] = df["Salary"].fillna(0)
df["Bonus"] = df["Bonus"].fillna(0)
print("\n->JOINED DATAFRAME AFTER HANDLING MISSING VALUES:\n",df)
df["FinalSalary"] = df["Salary"] + df["Bonus"]
print("\n->JOINED DATAFRAME AFTER VECTORIZATION:\n",df)
print("\n->Total Salary:", df["Salary"].sum())
print("\n->Average Salary:", df["Salary"].mean())
print("\n->Max FinalSalary:", df["FinalSalary"].max())
print("\n->Employee Count:", df["Name"].count())
df["Category"] = df["FinalSalary"].apply(
    lambda x: "High" if x >= 55000 else ("Medium" if x >= 45000 else "Low")
)
dept_map = {
    "IT": "Information Technology",
    "HR": "Human Resources",
    "Finance": "Finance Department"
}
emp_df2=emp_df.set_index("EmpID")
print("\n->EMPLOYEE DATAFRAME AFTER SETING EmpID AS INDEX:\n",emp_df2)
salary_df2 = salary_df.set_index("EmpID")
print("\n->SALARY DATAFRAME AFTER ADDING SALARY AS INDEX:\n",salary_df2)
joined_df = emp_df2.join(salary_df2)
print("\n->EMPLOYEE DATFRAME AFTER ADDING SALARY COLUMN  TO EMPLOYEE DATAFRAME:\n",joined_df)
jan_df = pd.DataFrame({
    "EmpID": [101, 102],
    "Sales": [2000, 3000]
})
feb_df = pd.DataFrame({
    "EmpID": [103, 104],
    "Sales": [2500, 3500]
})
monthly_report = pd.concat([jan_df, feb_df], axis=0)
print("\n->FINAL MONTHLY REPORT:\n",monthly_report)
```
