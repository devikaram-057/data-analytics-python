```python
import pandas as pd
data = {
    "EmpID": [101, 102, 103, 104, 105, 106],
    "Name": ["Asha", "Ravi", "Neha", "Arun", "Priya", "Kiran"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Gender": ["Female", "Male", "Female", "Male", "Female", "Male"],
    "Salary": [50000, 40000, 55000, 60000, 42000, 52000]
}
df = pd.DataFrame(data)
print("->Dataframe:\n",df)
grouped = df.groupby('Department')
print("\n->groupby() Basic:\n",grouped['Salary'].mean())
result = df.groupby("Department").agg({
    "Salary": ["sum", "mean","max","min"]
})
print("\n->Multi Aggregation:\n",result)
result = df.groupby("Department").agg(
    Total_Salary=("Salary","sum"),
    Average_Salary=("Salary","mean"),
    Employee_Count=("Salary","count")
)
print("\n->Named Aggregation:\n",result)
pivot = df.pivot_table(values='Salary', index='Department', aggfunc='mean')
print("\n->Pivot Table:\n",pivot)
ct = pd.crosstab(df['Gender'], df['Department'])
print("\n->Crosstab:\n",ct)
```
