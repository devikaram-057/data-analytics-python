```python
import pandas as pd
data = {
    "OrderID": [1, 2, 3, 4, 5, 6, 7],
    "Customer": ["Anu", "Rahul", "Meera", "Ajay", "Sara", "Vishnu", "Diya"],
    "City": ["Kochi", "Kochi", "Trivandrum", "Kochi", "Trivandrum", "Kochi", "Trivandrum"],
    "Category": ["Grocery", "Electronics", "Grocery", "Clothing", "Clothing", "Grocery", "Electronics"],
    "PaymentMode": ["UPI", "Card", "Cash", "UPI", "Card", "Cash", "UPI"],
    "Amount": [1200, 25000, 800, 3500, 4000, 950, 18000]
}

df = pd.DataFrame(data)
print(df)

grouped = df.groupby('Category')
print("\n->Grouping aggregation:\n",grouped['Category'].sum())
print(("\n->Multiple Column Grouping:\n",df.groupby(['City', 'Category'])['Amount'].mean()))
result = df.groupby("City").agg(
    Total_Sales=("Amount","sum"),
    Avg_Sales=("Amount","mean"),
    Order_Count=("OrderID","count")
)
print("\n->Named Aggregation:\n",result)
pivot = df.pivot_table(values='Amount', index='City', columns='PaymentMode', aggfunc='sum')
print("\n->Pivot Table:\n",pivot)
ct = pd.crosstab(df['Category'], df['PaymentMode'])
print("\n->Crosstab:\n",ct)
result = pd.melt(df, id_vars="City", var_name="PaymentMode", value_name="TotalAmount")
print("\n->Melt:\n",result)   
stacked =pivot.stack()
print("\n->Stacked:\n",stacked)
result=stacked.unstack()
print("\n->Unstacked:\n",result)
```
