```python
import pandas as pd
data = {
    "student_id": [1, 2, 3, 4, 5, 6],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "subject": ["Math", "Science", "English", "History", "Math", "Science"],
    "marks": [85, 90, 78, 88, 92, 80]
}

df = pd.DataFrame(data)
print(df)
df.to_csv("8Cstudents_details.csv", index=False, encoding="utf-8")
print("Exported to csv file")
df.to_excel("8Cstudents_details.xlsx", sheet_name="Marks_sheet", index=False)
print("Exported to Excel file")
df.to_json('8Cstudents_details.json')
print("Exported to JSON file")
```
