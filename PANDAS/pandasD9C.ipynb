```python
import pandas as pd
data = {
    "Date": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "Sales": [100, 150, 120, 170, 200]
}
df = pd.DataFrame(data)
dt = pd.to_datetime(df["Date"])
print(df,"\n",dt)
df["Year"] = dt.dt.year
df["Month"] =dt.dt.month
df["Day"] =dt.dt.day
print("\n->Extracted Year, Month, and Day from the Date column and create new columns:\n",df)
previous_day_sales = df["Sales"].shift(1)
print("\n->Previous_Day_Sales using shift(1):\n",previous_day_sales)
daily_change = df["Sales"].diff()
print("\n->Daily_Change using diff():\n",daily_change)
print("\n->Set the Date column as the index and retrieve sales for 2024-01-03 using indexing:\n",daily_change)
df = pd.DataFrame(data, index=df["Date"])
print(df.loc['2024-01-03'])
```
