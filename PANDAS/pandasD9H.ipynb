```python
import pandas as pd
import random
date_range = pd.date_range("2024-01-01", "2024-01-31")
random_dates = random.sample(list(date_range), 10)
revenues = [random.randint(1000, 10000) for _ in range(10)]
df = pd.DataFrame({
    "Date": random_dates,
    "Revenue": revenues
})
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(by="Date")
print("Initial DataFrame:\n", df)
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Previous_Day_Revenue"] = df["Revenue"].shift(1)
df["Revenue_Change"] = df["Revenue"].diff()
print("\nAfter calculations:\n", df)
df.set_index("Date", inplace=True)
print("\nAfter setting Date as index:\n", df)
weekly = df["Revenue"].resample('W').sum()
print("\nWeekly Revenue:\n", weekly)
monthly = df["Revenue"].resample('ME').sum()
print("\nMonthly Revenue:\n", monthly)
```
