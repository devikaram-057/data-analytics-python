```python
import pandas as pd
sales_series=pd.Series([250, 300, 150, 400, 350],index=["Mon", "Tue", "Wed", "Thu", "Fri"]
)
print("Sales Series:\n",sales_series)
print("Sales on wednesday:",sales_series["Wed"])
print("Total Sales:",sales_series.sum())
print("Average Sales:",sales_series.mean())
sales_df=pd.DataFrame({'Day':['Mon','Tue','Wed','Thu','Fri'],'Sales':[250, 300, 150, 400, 350]})
print("Sales DataFrame:\n",sales_df)
print("Maxinum sales:",sales_df['Sales'].max())
print("Mininum sales:",sales_df['Sales'].min())
```
