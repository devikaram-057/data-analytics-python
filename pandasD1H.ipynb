```python
import pandas as pd
student1_marks=pd.Series([85,78, 90, 88],index=["Math", "Science", "English", "Computer"])
print("Marks of student 1:\n",student1_marks)
print("Highest mark:",student1_marks.max())
print("Lowest mark:",student1_marks.min())
print("Total marks:",student1_marks.sum())
print("Average mark:",student1_marks.mean())
marks_df=pd.DataFrame({'Name':['Arun','Binu','Chithra'],
                       'Math':[80,70,92],
                       'Science':[78,82,88],
                       'English':[90,76,95],
                       'Computer':[88,80,91]})
print("Marks DataFrame:\n",marks_df)
marks_df['Total']=marks_df[['Math','Science','English','Computer']].sum(axis=1)
marks_df['Avg']=marks_df[['Math','Science','English','Computer']].mean(axis=1)
print("DataFrame with Total and Average columns:\n",marks_df)
print("Highest Total marks:",marks_df.loc[marks_df["Total"].idxmax(), "Name"])
print("Type of student1_marks:",type(student1_marks))
print("Type of marks_df:",type(marks_df))
```
