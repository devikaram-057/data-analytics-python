```python
import pandas as pd
data = {
    "EmpID": [201, 202, 203, 204, 205, 206],
    "Name": ["Arun", "Neha", "Sona", "Rahul", "Maya", "John"],
    "Age": [23, 29, 35, 27, 31, 26],
    "Dept": ["IT", "HR", "IT", "Finance", "IT", "HR"],
    "Salary": [25000, 32000, 45000, 38000, 50000, 28000],
    "Attendance": [78, 88, 92, 85, 95, 60],
    "Status": ["pending", "completed", "pending", "completed", "pending", "pending"],
    "City": ["Kochi", "Trivandrum", "New Delhi", "Mumbai", "New York", "Yorkshire"]
}
df=pd.DataFrame(data)
print(df)
print("\n->Print only Name and mark using []:")
print(df[['Name', 'Salary']])
print("\n->Print employee details of index 3 using .loc[]:\n",df.loc[3])
print("\n->Print employee details of 4th row using .iloc:\n",df.iloc[3])
print("\n->Select subset of rows 1 to 4 and columns Name, Dept, Attendance:\n",df.loc[1:5, ['Name', 'Dept','Attendance']])
print("->Employees with Salary > 35000:\n",df[df['Salary']>35000])
print("->Employees from IT department AND Attendance >= 90:\n")
result = df[(df["Dept"]=="IT") & (df["Attendance"]>=90)]
print(result)
print("\n->Employees where City contains York using .str.contains():\n",df[df["City"].str.contains("York")])
df["Bonus"] = df["Salary"]*0.10
df['FinalSalary']= df['Salary']+df['Bonus']
df.loc[df["Status"] == "pending", "Status"] = "in progress"
df = df.drop(columns=["Dept"])
bins = [0, 70, 85, 100]
labels = ["Low", "Average", "High"]
df["Performance"] = pd.cut(df['Attendance'], bins=bins, labels=labels)
print("\n->After categorizing attendance:\n",df[['Name','Attendance','Performance']])
df = df.map(lambda x: x*1)
print("\n->df after applying transformations:\n",df)
```
